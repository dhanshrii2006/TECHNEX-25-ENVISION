```python
import pandas as pd

# Load the dataset
data_path = "heart_disease.csv"  # Replace with the correct file path
df = pd.read_csv(data_path)

# Display the first few rows to understand the dataset
print(df.head())

# Check for missing values and data types
print(df.info())

# Basic statistics of the dataset
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribution of Cholesterol Levels
plt.figure(figsize=(8, 5))
sns.histplot(df['chol'], kde=True, color='green')
plt.title("Distribution of Cholesterol Levels")
plt.xlabel("Cholesterol")
plt.ylabel("Frequency")
plt.savefig("cholesterol_distribution.png")
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 2. Age vs Cholesterol (colored by Target)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='chol', hue='target', data=df, palette='coolwarm')
plt.title("Age vs Cholesterol (Colored by Target)")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.savefig("age_vs_chol.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 3. Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("correlation_heatmap.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 4. Maximum Heart Rate Across Target
plt.figure(figsize=(8, 5))
sns.boxplot(x='target', y='thalach', data=df, palette='Set2')
plt.title("Maximum Heart Rate Across Target")
plt.xlabel("Target (0 = No Disease, 1 = Disease)")
plt.ylabel("Maximum Heart Rate")
plt.savefig("max_heart_rate_target.png")
plt.show()
```
