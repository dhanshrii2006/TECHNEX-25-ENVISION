```python
import pandas as pd

# Load the dataset
data_path = "diabetes.csv"  # Adjust this path if necessary
df = pd.read_csv(data_path)

# Display the first few rows
print(df.head())

# Check for missing values
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of a numerical column (e.g., Glucose levels)
plt.figure(figsize=(8, 5))
sns.histplot(df['Glucose'], kde=True, color='blue')
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.savefig("distribution_glucose.png")  # Save the plot
plt.show()




```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot to compare BMI across Diabetes Outcomes
plt.figure(figsize=(8, 5))
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title("BMI vs Diabetes Outcome")
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("BMI")
plt.savefig("bmi_vs_outcome.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("correlation_heatmap.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Scatter plot for Age vs Glucose
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Glucose', hue='Outcome', data=df)
plt.title("Age vs Glucose Levels")
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.savefig("age_vs_glucose.png")
plt.show()
```
